```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime, timedelta

import os
from glob import glob
```

```python
path = os.getcwd()

file = f"/masters_diploma/processed_data/history_weather.csv"
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
```

```python
df = pd.read_csv(
    file, 
    parse_dates=['time'],
    date_parser=dateparse,
    index_col=['time']
)
df.index.name = 'date'
```

```python
df
```

```python
from sklearn.ensemble import RandomForestRegressor

import xgboost
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor(**{
    'n_estimators': 500,
    'n_jobs': -1,
    'max_depth': 6,
    'eta': 0.3,
    'booster': 'gbtree',
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'subsample': 1,
    'colsample_bytree': 1,
    'min_child_weight': 1,
    'random_state': 7,
})
```

```python
# X_train = df.loc[:'2023-10-01', list(set(df.columns)-set(['close_price', 'adj_close_price']))]
X_train = df.loc[:'2024-12-31', ['relative_humidity', 'dew_point', 'weather_code', 'pressure_msl', 
                                   'is_day', 'sunshine_duration', 'month', 'year_day']]
y_train = df.loc[:'2024-12-31', ['temperature']]

# X_test = df.loc['2023-10-01':, list(set(df.columns)-set(['close_price', 'adj_close_price']))]
X_test = df.loc['2024-12-31':, ['relative_humidity', 'dew_point', 'weather_code', 'pressure_msl', 
                                  'is_day', 'sunshine_duration', 'month', 'year_day']]
y_test = df.loc['2024-12-31':, ['temperature']]

y_pred = model_xgb.fit(X_train, y_train).predict(X_test)
```

```python
y_pred_df = pd.DataFrame(y_pred, index=y_test.index)

plt.plot(df.loc['2024-12-31':, ['temperature']])
plt.plot(y_pred_df)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
err = pd.DataFrame({
    'abs_err': abs(y_pred_df[0] - y_test['temperature']),
    'rel_err': abs(y_pred_df[0] / y_test['temperature'] - 1) * 100
})
err.mean()
```
